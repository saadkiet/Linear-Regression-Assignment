```python
#importing Dataset
from sklearn.datasets import load_boston
boston = load_boston()
print(boston)
```

```python
# making Dataframe for independent and dependent variables for the dataset
df_x=pd.DataFrame(data=boston.data, columns=boston.feature_names)
df_x.head()
```

```python
# checking for NULL values
df_x.isnull().sum()
```

```python
#checking stats of the Independent variables
df_x.describe()
```

```python
# Standardizinng the values in the dataset
from sklearn.preprocessing import StandardScaler
scaler =StandardScaler()
X_scaled = scaler.fit_transform(df_x)
```

```python
# checking multicolinearity

from statsmodels.stats.outliers_influence import variance_inflation_factor
variables = X_scaled

vif = pd.DataFrame()


vif["VIF"] = [variance_inflation_factor(variables, i) for i in range(variables.shape[1])]


vif["Features"] = df_x.columns
```

```python
vif
```

#### Here we can see that column RAD and TAX have Variance inflation factor score of above 5. They have very high multicollinearity. So we will drop them, as it will affact our predictions

```python
X_scaled_updated= pd.DataFrame(X_scaled, columns=df_x.columns)
X_scaled_updated.head()
```

```python
X_scaled_updated.drop(['RAD','TAX'],axis=1,inplace=True)
X_scaled_updated.head()
```

```python
col_=["price"]
df_y=pd.DataFrame(data=boston.target,columns=col_)
df_y.head()
```

```python
## Importing Linear Regression Model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
# Splitting dataset intro train and test
from sklearn.model_selection import train_test_split

xtrain,xtest, ytrain, ytest= train_test_split(X_scaled_updated,df_y,test_size=0.3, random_state=42)
```

```python
# Model training
#Train our model with the training data
reg.fit(xtrain, ytrain)
```

```python
# Let's create a function to create adjusted R-Squared
def adj_r2(x,y):
    r2 = reg.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adjusted_r2 = 1-(1-r2)*(n-1)/(n-p-1)
    return adjusted_r2
```
